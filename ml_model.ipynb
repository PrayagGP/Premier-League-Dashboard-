```python
import numpy as np

# Define a function to map predicted values to FTR categories
def map_to_FTR_category(pred):
    if pred <= 0.75:
        return 'D'
    elif pred <= 1.25:
        return 'A'
    else:
        return 'H'

# Predict FTR for missing_df
ftr_pred = f_model.predict(missing_df[predictors])

# Map predicted values to FTR categories using the defined function
ftr_labels_mapped = [map_to_FTR_category(pred) for pred in ftr_pred]

# Add the mapped predicted FTR values to missing_df
missing_df['Predicted_FTR_Mapped'] = ftr_labels_mapped

# Display the predictions
print(missing_df[['HomeTeam', 'AwayTeam', 'Predicted_FTR_Mapped']])
```

```python
def kiran_xgb(n, e, g, m, s):
    h_model = xgb.XGBRFRegressor(n_estimators = n, eta=e, gamma=g, max_depth=m, subsample=s)
    h_model = h_model.fit(train[predictors], train["FTHG"])
    h_pred = h_model.predict(test[predictors])
    a_model = xgb.XGBRFRegressor(n_estimators = n,eta=e, gamma=g, max_depth=m, subsample=s)
    a_model = a_model.fit(train[predictors], train["FTAG"])
    a_pred = a_model.predict(test[predictors])
    preds_FTR = ['D' if abs(pred_home - pred_away) <= 0.2 else 'H' if pred_home > pred_away else 'A' for pred_home, pred_away in zip(h_pred, a_pred)]
    acc_FTR = accuracy_score(test["FTR"], preds_FTR)
    print(acc_FTR, n, e, g, m, s)
```

```python
e = 0.3
n = 20
g = 1
m = 18
s = 0.6
kiran_xgb(n, e, g, m, s)
```

```python
predicted_fthg_list = []
predicted_ftag_list = []
predicted_ftr_list = []

for index, row in missing_df.iterrows():
    predictors_missing = row[predictors].values.reshape(1, -1)

    fthg_pred = h_model.predict(predictors_missing)[0]

    ftag_pred = a_model.predict(predictors_missing)[0]

    ftr_pred = 'H' if fthg_pred > ftag_pred else 'A' if fthg_pred < ftag_pred else 'D'

    predicted_fthg_list.append(fthg_pred)
    predicted_ftag_list.append(ftag_pred)
    predicted_ftr_list.append(ftr_pred)

missing_df['Predicted_FTHG'] = predicted_fthg_list
missing_df['Predicted_FTAG'] = predicted_ftag_list
missing_df['Predicted_FTR'] = predicted_ftr_list

print(missing_df[['Predicted_FTHG', 'Predicted_FTAG', 'Predicted_FTR']])
```
